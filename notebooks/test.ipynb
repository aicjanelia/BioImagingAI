```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#| label: fig-test-fig
#| code-fold: true
#| fig-cap: "Polar plot of circles of random areas at random coords"
#| fig-alt: a demo figure to be embedded in a template chapter
np.random.seed(42)

N = 100
r = 2*np.random.rand(N)
theta = 2*np.pi * np.random.rand(N)
area = 100 * r**2
colors = theta

fig = plt.figure()
ax  = fig.add_subplot(projection='polar')
c = ax.scatter(theta,r,c=colors,s=area,cmap='plasma',alpha=0.75)
```
